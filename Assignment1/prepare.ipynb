#### Importing libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
import requests
import zipfile
from google.colab import drive
drive.mount('/content/drive')
```

```python
def download_and_extract(url, output_path):
    """Downloads and extracts a dataset from a given URL."""
    zip_path = os.path.join(output_path, "sms_spam.zip")
    os.makedirs(output_path, exist_ok=True)

    response = requests.get(url)
    with open(zip_path, "wb") as f:
        f.write(response.content)

    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(output_path)

    os.remove(zip_path)  # Cleanup after extraction

def load_data(file_path):
    """Loads the SMS Spam Collection dataset from a given file path."""
    df = pd.read_csv(file_path, sep='\t', header=None, names=['label', 'message'])
    return df

def preprocess_data(df):
    """Preprocesses the dataset (if needed)."""
    df['label'] = df['label'].map({'ham': 0, 'spam': 1})  # Convert labels to numerical
    return df

def split_and_store_data(df, output_dir):
    """Splits the data into train, validation, and test sets and saves them as CSV files."""
    train, temp = train_test_split(df, test_size=0.3, random_state=42, stratify=df['label'])
    val, test = train_test_split(temp, test_size=0.5, random_state=42, stratify=temp['label'])

    os.makedirs(output_dir, exist_ok=True)
    train.to_csv(os.path.join(output_dir, 'train.csv'), index=False)
    val.to_csv(os.path.join(output_dir, 'validation.csv'), index=False)
    test.to_csv(os.path.join(output_dir, 'test.csv'), index=False)
```

```python
if __name__ == "__main__":
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00228/smsspamcollection.zip"
    data_dir = "/content/drive/MyDrive/Applied_ML/data"
    download_and_extract(url, data_dir)

    file_path = os.path.join(data_dir, "SMSSpamCollection")
    df = load_data(file_path)
    df = preprocess_data(df)
    split_and_store_data(df, data_dir)

    print("Data prepared and stored successfully.")
```
